```python
from FlightRadar24.api import FlightRadar24API
import pandas as pd
fr_api = FlightRadar24API()
```

```python
airports = fr_api.get_airports()
selesai = pd.DataFrame(airports)
```

```python
selesai.to_excel("excelflight.xlsx")
selesai.to_csv("csvflight.csv")
```

```python
df = pd.read_csv('csvflight.csv')
```
